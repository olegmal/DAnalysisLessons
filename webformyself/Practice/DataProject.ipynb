1.Конкатенация разных файлов в один датасет:

```python
import pandas as pd
```

```python
import os
```

```python
df = pd.DataFrame()
for f in os.listdir('files/'):
    tmp = pd.read_csv('files/' + f)
    df = pd.concat([df, tmp])
```

```python
df
```

```python
df.to_csv('result.csv', index=False)
```

Добавление колонки Month

```python
df = pd.read_csv('result.csv')
df
```

```python
df.info(memory_usage='deep')
```

```python
df['Order_Date'] = pd.to_datetime(df.Order_Date, format='%m/%d/%y %H:%M')
df
```

```python
df.info(memory_usage='deep')
```

```python
df['Month'] = df.Order_Date.dt.month
df
```

```python
df.to_csv('result.csv', index=False)
```

Определяем лучший месяц продаж

Группируем по месяцам и смотрим Total

```python
res = df.groupby('Month')[['Total']].sum()
res
```

```python
res.plot(kind='bar', grid=True, ylabel='USD (mln.)')
```
