```python
import os
import pandas as pd
```

```python
df = pd.read_csv('result.csv')
df
```

```python
df.info(memory_usage='deep')
```

```python
res = df.groupby('Month')[['Total']].sum()
res
```

```python
res.plot(kind='bar', grid=True, ylabel='USD (mln.)')
```

Чтобы убрать scientidic format (le6):

```python
ax = res.plot(kind='bar', grid=True)
ax.set_ylabel('USD (mln.)')
ax.get_yaxis().get_major_formatter().set_scientific(False)
```

```python
plt.bar(res.index, res.Total)
plt.xticks(range(1, 13))
plt.yticks(range(0, 5000000, 500000))
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.grid()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [8, 6]
```

```python
plt.bar(res.index, res.Total)
plt.xticks(range(1, 13))
plt.yticks(range(0, 5000000, 500000))
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.xlabel('Months')
plt.ylabel('Income in $')

for index, value in enumerate(res.Total):
    plt.text(index + 1,
             500000,
             round(value),
             rotation='vertical',
             fontsize='12',
             color='white',
             ha='center')
# plt.grid()
plt.show()
```
