```python
import pandas as pd
```

```python
df = pd.read_csv("D:/PythonProject/webformyself/country.csv", sep=";")
df
```

```python
df['cnt']=5
df
```

```python
df.cnt = df.cnt.apply(lambda x: 1 if x > 100 else 2)
df
```

```python
df.drop(columns='cnt', inplace=True)
df
```

```python
df['cnt'] = df.Name.apply(len)
df
```

```python
def get_size(param):
    if param > 10000000:
        return 'mega'
    elif param > 1000000:
        return 'large'
    elif param > 500000:
        return 'medium'
    else:
        return 'small'
    
df['Size'] = df.Population.apply(get_size)        
df
```

```python
df.Size.value_counts()
```

```python
df[df.Size== 'mega']
```

```python
df['Size2']= df.Population.apply(lambda x: 'mega' if x > 100000000 else 'large' if x > 1000000 else 'medium' if x > 500000 else 'small')
df
```

```python
df.Size2.value_counts()
```

```python
df.drop(columns=['Size', 'Size2'], inplace=True)
df
```

```python
df.drop(columns='cnt', inplace=True)
df
```

```python
def get_density(row):
    return int(row['Population']/row['SurfaceArea'])
df['Density'] = df.apply(get_density, axis=1)
df
```

```python
for index, row in df.iterrows():
    print(index, row.Name, row.Density, sep=' | ')
```
