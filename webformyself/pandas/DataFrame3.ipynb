```python
df = pd.read_csv("D:\PythonProject\webformyself\price1.csv", sep=';')
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df[df.price.isnull()]
```

```python
df.loc[df.qty.isnull(), 'id': 'price']
```

```python
df.dropna(how='any')
```

```python
# удаление строк, где ВСЕ ячейки в строке имеют NaN
df.dropna(how='all')
```

```python
# удаление заданных строк с определением колонок
df.dropna(how='all', subset=['price', 'qty'], inplace=True)
```

```python
# замена значений в ячейках с NaN
df.fillna(value=0)
```

```python
df.fillna(value={'id': '-', 'title': '-', 'price': 0, 'qty': 0})
```

```python
# заполение пустіх ячеек предідущими значениями 
df.ffill()
```

```python
# заполение пустіх ячеек последующими значениями 
df.bfill()
```
