```python
df = pd.read_csv("D:/PythonProject/webformyself/21_files/price.csv", sep=';')
df
```

```python
df.title.str.extract('(?P<note>\w+) (?P<brand>\w+)')
```

```python
df['brand'] = df.title.str.extract('\w+ (?P<brand>\w+)')
df
```

```python
df.shape
```

```python
len(df.brand.unique())
```

```python
df.brand.nunique()
```

```python
df.brand.unique()
```

```python
brands = {}
for el in enumerate(df.brand.unique(), 1):
    brands[el[1]] = el[0]
brands
```

```python
df['brand_id'] = df.brand.apply(lambda x: brands[x])
df
```

```python
df_brands = df.loc[:, ['brand_id', 'brand']].drop_duplicates()
df_brands
```

```python
df.drop(columns=['brand'], inplace=True)
```

```python
df
```

```python
df.to_csv('products.csv', index=False, sep=';')
```

```python
df_brands.to_csv('brands.csv', index=False, sep=';')
```
